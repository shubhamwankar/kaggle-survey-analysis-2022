# Reading and Cleaning Data

```python
data = pd.read_csv('../input/kaggle-survey-2022/kaggle_survey_2022_responses.csv')
```

```python
data.head(2)
```

```python
# Correcting the column names
questions = data.iloc[0,1:].values.tolist()
question_num = data.iloc[0,1:].index.tolist()
header_str = [question_num[i] + " " + questions[i] for i in range(len(questions))]
header_str.insert(0, data.iloc[0,0])
data.columns = header_str
data = data.drop(0, axis = 0)
data.head(2)
```

```python
# Finding questions with choices / multiple choices
select_questions = []
for header in header_str:
    if 'select' in header.lower():
        select_questions.append(header)

# Filtering the questions to their unique question
unique_select_questions = []
question_numbers = []
for col in select_questions[1:]:
    if '_' in col:
        question = col.split('(S')[0]
        question_number = question.split('_')[0]
        question = question.split()[1:]
        question = " ".join(question)
        if question not in unique_select_questions:
            unique_select_questions.append(question)
        if question_number not in question_numbers:
            question_numbers.append(question_number)

unique_select_questions_cleaned = [question_numbers[i] + \
                                  " " + \
                                  unique_select_questions[i] \
                                   for i in range(len(question_numbers))]
```

## Algorithm to clean this data
Initialize an empty dataframe  

For each question with choice / multiple choices:  

&nbsp;&nbsp;&nbsp;&nbsp;> Filter all the subquestions / choices.  

&nbsp;&nbsp;&nbsp;&nbsp;> Filter the dataframe with subquestions / choices and store the answers values in a list  

&nbsp;&nbsp;&nbsp;&nbsp;> For each answser in the list of answers:  

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;> For each element in the answer:  

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;> If the element is not NaN then append the element to a temporary list.  

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;> Try to join the elements for temporary list using join function  

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;> If there's a exception then try to convert the element into a string and then join them  

&nbsp;&nbsp;&nbsp;&nbsp;> Store the cleaned answer values with respect to each question with a choice / multiple choices in a dataframe.

```python
# Applying the above algorithm to clean the data
cleaned_df = pd.DataFrame()
for i in range(len(unique_select_questions)):
    sub_questions = [col for col in data.columns if unique_select_questions[i] in col]
    answers = data[sub_questions].values.tolist()
    cleaned = []
    for answer in answers:
        temp = []
        for element in answer:
            if element is not np.nan:
                temp.append(element)
        try:
            temp = ", ".join(temp)
            cleaned.append(temp)
        except:
            temp = [str(j) for j in temp]
            temp = ", ".join(temp)
            cleaned.append(temp)
    cleaned_df[unique_select_questions[i]] = cleaned
```

```python
# Naming the columns correctly
cleaned_df.columns = unique_select_questions_cleaned
```

```python
# Get column names for all questions without multiple sub questions / choices
unique_questions = []
for col in data.columns:
    if '_' not in col:
        unique_questions.append(col)
```

```python
# Concatenate the two dataframe
unique_df = data[unique_questions]
final_df = pd.concat([cleaned_df, unique_df], axis = 1)
final_df.head(2)
```

```python
# Replace "" with None
final_df = final_df.replace('', 'None')
```

```python
# Store all the column names to a list
columns = final_df.columns.tolist()
```

### Survey Response Time Statstics

```python
# Converting the seconds into float value and plotting statistics of the Response Time
final_df['Duration (in seconds)'] = final_df['Duration (in seconds)'].apply(lambda x: float(x))
median = final_df['Duration (in seconds)'].median()
final_df['Duration (in seconds)'] = final_df['Duration (in seconds)'].fillna(median)
duration = final_df['Duration (in seconds)'].apply(lambda x: round((x / 60), 2))
duration.name  = 'Duration (in minutes)'
stats = duration.describe().to_frame()
stats = stats['Duration (in minutes)'].apply(lambda x: int(x))
stats = stats.drop(['std', '25%', '75%'], axis=0)
stats.index = ['Total Number of Responses', 'Mean Response Time (in minutes)',
              'Fastest Response Time (in minutes)', 'Median Response Time (in minutes)',
              'Longest Response Time (in minutes)']
fig = px.scatter(stats, x=stats.index, y='Duration (in minutes)',
                 text='Duration (in minutes)', size='Duration (in minutes)',
                 title='Survey Response Time Statistics')
fig.update_traces(hovertemplate='%{x} <br>%{y}')
fig.update_layout(font_size=14)
fig.update_xaxes(title='')
fig.update_yaxes(title='')
fig.show()
```

```python
def clean_dict(d):
    init = 0
    cntr = 0
    name = ''
    new_name = ''
    new_dict = {}
    for key, val in d.items():
        if (init == val) and (cntr != 1):
            cntr = 1
            new_name = name + ", " + key
            new_dict[new_name] = new_dict.pop(name)
        elif (cntr == 1) and (val == init):
            temp = new_name + ", " + key
            new_dict[temp] = new_dict[new_name]
            del new_dict[new_name]
            new_name = temp
        elif (cntr == 1) and (val != init):
            cntr = 0
            new_dict[new_name] = init
        else:
            name = key
            init = val
            new_dict[name] = val

    return new_dict
```

```python
def get_unique_vals(final_df, col=''):
    vals = final_df[col].values.tolist()
    val_dict = {}
    for element in vals:
        try:
            keys = element.split(', ')    
        except:
            pass
        for key in keys:
            key = key.strip()
            if key in val_dict:
                val_dict[key] += 1
            else:
                val_dict[key] = 1

    return val_dict
```
